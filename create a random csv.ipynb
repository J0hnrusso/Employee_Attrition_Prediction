```python
import csv
import random

# Define the headers
headers = [
    "Age", "Attrition", "DailyRate", "DistanceFromHome", "EmployeeNumber", "EnvironmentSatisfaction",
    "JobInvolvement", "JobLevel", "JobSatisfaction", "MonthlyIncome", "OverTime", "StockOptionLevel",
    "TotalWorkingYears", "TrainingTimesLastYear", "WorkLifeBalance", "YearsAtCompany", "YearsInCurrentRole",
    "YearsWithCurrManager", "BusinessTravel_Travel_Frequently", "BusinessTravel_Travel_Rarely",
    "MaritalStatus_Married", "MaritalStatus_Single", "EducationField_Life Sciences", "EducationField_Marketing",
    "EducationField_Medical", "EducationField_Other", "EducationField_Technical Degree", "JobRole_Human Resources",
    "JobRole_Laboratory Technician", "JobRole_Manager", "JobRole_Manufacturing Director", "JobRole_Research Director",
    "JobRole_Research Scientist", "JobRole_Sales Executive", "JobRole_Sales Representative"
]

# Function to generate random data
def generate_random_row(employee_number):
    age = random.randint(20, 65)
    attrition = random.choice([True, False])
    daily_rate = random.randint(200, 2000)
    distance_from_home = random.randint(1, 30)
    environment_satisfaction = random.randint(1, 4)
    job_involvement = random.randint(1, 4)
    job_level = random.randint(1, 5)
    job_satisfaction = random.randint(1, 4)
    monthly_income = random.randint(1000, 20000)
    over_time = random.choice([True, False])
    stock_option_level = random.randint(0, 3)
    total_working_years = random.randint(0, 40)
    training_times_last_year = random.randint(0, 6)
    work_life_balance = random.randint(1, 4)
    years_at_company = random.randint(0, 40)
    years_in_current_role = random.randint(0, 20)
    years_with_curr_manager = random.randint(0, 20)
    
    # Business Travel (mutually exclusive)
    business_travel_freq = random.choice([True, False])
    business_travel_rare = not business_travel_freq if business_travel_freq else random.choice([True, False])
    
    # Marital Status (mutually exclusive)
    marital_status_married = random.choice([True, False])
    marital_status_single = not marital_status_married
    
    # Education Field (one-hot encoded)
    education_fields = ["Life Sciences", "Marketing", "Medical", "Other", "Technical Degree"]
    education_field = random.choice(education_fields)
    education_field_life_sciences = education_field == "Life Sciences"
    education_field_marketing = education_field == "Marketing"
    education_field_medical = education_field == "Medical"
    education_field_other = education_field == "Other"
    education_field_technical_degree = education_field == "Technical Degree"
    
    # Job Role (one-hot encoded)
    job_roles = [
        "Human Resources", "Laboratory Technician", "Manager", "Manufacturing Director",
        "Research Director", "Research Scientist", "Sales Executive", "Sales Representative"
    ]
    job_role = random.choice(job_roles)
    job_role_hr = job_role == "Human Resources"
    job_role_lab_tech = job_role == "Laboratory Technician"
    job_role_manager = job_role == "Manager"
    job_role_mfg_dir = job_role == "Manufacturing Director"
    job_role_research_dir = job_role == "Research Director"
    job_role_research_sci = job_role == "Research Scientist"
    job_role_sales_exec = job_role == "Sales Executive"
    job_role_sales_rep = job_role == "Sales Representative"
    
    return [
        age, attrition, daily_rate, distance_from_home, employee_number, environment_satisfaction,
        job_involvement, job_level, job_satisfaction, monthly_income, over_time, stock_option_level,
        total_working_years, training_times_last_year, work_life_balance, years_at_company,
        years_in_current_role, years_with_curr_manager, business_travel_freq, business_travel_rare,
        marital_status_married, marital_status_single, education_field_life_sciences, education_field_marketing,
        education_field_medical, education_field_other, education_field_technical_degree, job_role_hr,
        job_role_lab_tech, job_role_manager, job_role_mfg_dir, job_role_research_dir, job_role_research_sci,
        job_role_sales_exec, job_role_sales_rep
    ]

data = [headers]
for i in range(1, 1001):
    data.append(generate_random_row(i))

# Write to CSV
with open("random_employee_data_1000.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file 'random_employee_data.csv' created successfully!")
```
